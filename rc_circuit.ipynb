```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
voltage = pd.read_csv("/Users/jamesc/Downloads/voltage_4.7k_33.txt", header=None, names=["time", "voltage"])
```

```python
plt.plot(voltage.time, voltage.voltage);
```

```python
tmin, tmax = 2.1, 3
```

```python
time_charge = voltage.time[np.logical_and(voltage.time >= tmin, voltage.time <= tmax)]
voltage_charge = voltage.voltage[np.logical_and(voltage.time >= tmin, voltage.time <= tmax)]
plt.plot(time_charge, voltage_charge);
```

```python
exponent = -np.log(1 - voltage_charge/(voltage_charge.max() + 10e-6))
```

```python
plt.plot(time_charge, exponent);
```

```python
tmin, tmax = 2, 2.3
```

```python
time_charge_2 = time_charge[np.logical_and(time_charge >= tmin, time_charge <= tmax)]
exponent_charge_2 = exponent[np.logical_and(time_charge >= tmin, time_charge <= tmax)]
```

```python
plt.plot(time_charge_2, exponent_charge_2);
```

```python
time_sk = time_charge_2.values.reshape(-1, 1)
exponent_sk = exponent_charge_2.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(time_sk, exponent_sk)
```

Machine learned answer:

```python
1/reg.coef_
```

```python
4700*33e-6
```
